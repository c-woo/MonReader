```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import tensorflow as tf
from skimage.io import imread
from skimage.transform import resize, rotate, AffineTransform, warp
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from keras.callbacks import EarlyStopping
```

I will be attempting to create a neural network that is able to determine whether or not a page is being flipped from a series of images.

```python
## create variables for image locations on computer

train = './images/training'
train_f = './images/training/flip'
train_nf = './images/training/notflip'

test = './images/testing'
test_f = './images/testing/flip'
test_nf = './images/testing/notflip'
```

```python
## find total number of training and testing pictures

print('Number of Training flip pictures: ', len(os.listdir(train_f)))
print('Number of Training not flip pictures: ', len(os.listdir(train_nf)))
print('Number of Testing flip pictures: ', len(os.listdir(test_f)))
print('Number of Testing not flip pictures: ', len(os.listdir(test_nf)))
```

```python
## get image vectors from image files and append them to a list

train_flip = []
train_notflip = []
test_flip = []
test_notflip = []

for img in os.listdir(train_f):
    train_flip.append(imread(f'./images/training/flip/{img}'))
    
for img in os.listdir(train_nf):
    train_notflip.append(imread(f'./images/training/notflip/{img}'))
                                
for img in os.listdir(test_f):
    test_flip.append(imread(f'./images/testing/flip/{img}'))
    
for img in os.listdir(test_nf):
    test_notflip.append(imread(f'./images/testing/notflip/{img}'))
```

## Image Augmentation

I'll be doing image augmentation on one picture before applying the transformations to the training set.

```python
## use a training picture to test transformation on

testpic = train_flip[0]
plt.imshow(testpic)
```

Most of the pictures have a lot of dead space on the bottom below the book so I'll be cropping the pictures to take out most of the space without the book.

Divide the image array by 255 so each pixel is a value between 0 and 1 instead of 0 and 255.

```python
## test resizing image

plt.imshow(resize(cropped, (140, 100)))
```

```python
## rotate the images by a random degree between 0 and 180

plt.imshow(rotate(cropped, random.randint(0, 180)))
```

```python
afine_tf = AffineTransform(shear=0.1)
shear = warp(cropped, inverse_map=afine_tf)
plt.imshow(shear)
```

```python
train_flip = [resize(x[100:1550]/255, (140, 100)) for x in train_flip]
train_notflip = [resize(x[100:1550]/255, (140, 100)) for x in train_notflip]
```

```python
train_flip = [rotate(x, random.randint(0, 180)) for x in train_flip]
train_notflip = [rotate(x, random.randint(0, 180)) for x in train_notflip]
```

```python
##
##train_flip = [warp(x, inverse_map=afine_tf) for x in train_flip]
##train_notflip = [warp(x, inverse_map=afine_tf) for x in train_notflip]
```

```python
test_flip = [resize(x, (140, 100)) for x in test_flip]
test_notflip = [resize(x, (140, 100)) for x in test_notflip]
```

```python
training_data = []
testing_data = []

for label in [train_notflip, train_flip]:
    for img in label:
        if label == train_notflip:
            training_data.append([img, 0])
        else:
            training_data.append([img, 1])

for label in [test_notflip, test_flip]:
    for img in label:
        if label == test_notflip:
            testing_data.append([img, 0])
        else:
            testing_data.append([img, 1])
```

```python
random.shuffle(training_data)
random.shuffle(testing_data)
```

```python
X_train, X_test, y_train, y_test = [], [], [], []

for features, label in training_data:
    X_train.append(features)
    y_train.append(label)
    
for features, label in testing_data:
    X_test.append(features)
    y_test.append(label)
    
X_train = np.array(X_train)
X_test = np.array(X_test)
    
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (2, 2), activation = 'relu', padding = 'same', input_shape=(140, 100, 3)))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Conv2D(64, (2, 2), activation = 'relu', padding = 'same'))
model.add(layers.MaxPooling2D((2, 2)))

model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.Adam(lr=0.001), metrics=tf.keras.metrics.AUC(name="auc"))
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', patience=5, verbose=1)
```

```python
hist = model.fit(X_train, y_train, epochs=20, validation_split=0.25, callbacks=[es])
```

```python
# plot training loss and auc

plt.figure(figsize=(20, 8))
plt.suptitle('Loss and Accuracy Plots', fontsize=20)

plt.subplot(1, 2, 1)
plt.plot(hist.history['loss'], label='training loss')
plt.plot(hist.history['val_loss'], label='validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(hist.history['auc'], label='training auc')
plt.plot(hist.history['val_auc'], label='validation auc')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, y_test)
```
